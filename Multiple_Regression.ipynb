```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
```

```python
df = pd.read_csv("FuelConsumption.csv")
df.head()
```

```python
cdf = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB','CO2EMISSIONS']]
```

Testing and Training data splits using k-fold cross validation

```python
from sklearn.model_selection import KFold
from sklearn import linear_model
regr = linear_model.LinearRegression()

train_x = np.asanyarray(cdf[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']])
train_y = np.asanyarray(cdf[['CO2EMISSIONS']])

kf = KFold(n_splits=2)
kf.get_n_splits(train_x)

for train_index , test_index in kf.split(train_x):
  print( "TRAIN:", train_index , "TEST:", test_index)
  X_train, X_test = train_x[train_index], train_x[test_index]
  Y_train , Y_test = train_y[train_index] , train_y[test_index]

regr.fit(X_train, Y_train) 
```

```python
print('Coefficients :' , regr.coef_)
print('Intercept:',  regr.intercept_)
```

```python
from sklearn.metrics import r2_score
test_y = regr.predict(X_test)
print("R2- Score: %.2f" % r2_score(Y_test, test_y))
```
